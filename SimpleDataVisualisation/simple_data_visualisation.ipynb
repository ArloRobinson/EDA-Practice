# Simple Data Visualisation
Visualising data from `csv` files using a variety of graphs including 
- Histogram
- Stem and Leaf Plot
- Boxplot
- Scatterplot
- Descriptive Statistics

```python
# Imports 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import stemgraphic as stg
```

## Histogram
Creating a Histogram graph with data from the `restaurant.csv` file.

```python
# Extract restaurant.csv data with Pandas
hist_data = pd.read_csv("restaurant.csv", sep = ',', header = 0)
hist_data.head()
```

```python
sns.set_theme(style="whitegrid")
sns.histplot(data = hist_data, x = 'Price', bins = 11, binwidth = 5, kde = True, color = 'darkblue')
```

## Stem and Leaf Plot
Creating a Stem and Leaf plot with data from the `late_payment.csv` file.

```python
# Extract late_payment.csv data with Numpy
stem_data = np.genfromtxt("late_payment.csv", delimiter = ',', skip_header = 1)
stem_data
```

```python
fig, ax = stg.stem_graphic(stem_data)
```

## Boxplot
Creating a Boxplot with data from the `late_payment.csv` file.

```python
plot = sns.boxplot(x=stem_data, color="lightblue")
```

## Regplot
Creating a Regplot with data from the `property.csv` file.

```python
# Extract property.csv data with Pandas
scat_data = pd.read_csv("property.csv", sep = ',', header = 0)
scat_data.head()
```

```python
#sns.scatterplot(data = scat_data, x = 'Bedrooms', y = 'Asking_Price')
sns.regplot(data = scat_data, x = 'Bedrooms', y = 'Asking_Price')
```
